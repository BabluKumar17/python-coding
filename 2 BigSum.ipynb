```python
def bigSum(alist):
    return sum(i for i in alist)
```

```python
alist = list(map(int, input().strip().split(' ')))
result = bigSum(alist)
print(result)
```

```python
# Alternatively:

from functools import reduce

def bigSum(alist):
    return reduce(lambda x,y: x+y, alist)


print(bigSum([1000000001, 1000000002, 1000000003, 1000000004, 1000000005]))
```
