Given a square matrix of size $N*N$, calculate the absolute difference between the sums of its diagonals.

```python
matrix = [[11, 2, 4],
          [ 4, 5, 6],
          [10, 8,-12]]
```

```python
# Taking inputs

# number of rows
n = int(input().strip())
a = []
# go through as many times user wants
for a_i in range(n):
    a_t = [int(a_temp) for a_temp in input().strip().split(' ')]
    # and append it to "a"
    a.append(a_t)
```

```python
matrix
```

```python
def diagonal_difference(matrix):
    n = 0 
    m = -1
    diagonal1, diagonal2 = 0, 0
    for i, e in enumerate(matrix):
        diagonal1 += matrix[i][n]
        #print(matrix[i][n])
        diagonal2 += matrix[i][m]
        #print(matrix[i][m])
        n += 1
        m -= 1
    return abs(diagonal1-diagonal2)
            
print(diagonal_difference(matrix))
print(diagonal_difference(a))
```
