<center> <h1>Compare the Triplets </h1></center>

```python
def solve(a_triplet, b_triplet):
    alice_points = 0
    bob_points = 0
    for a_val, b_val in zip(a_triplet, b_triplet):
        if a_val < b_val:
            bob_points += 1
        elif a_val > b_val:
            alice_points += 1
    return (alice_points, bob_points)


a_triplet = map(int, input().split())
b_triplet = map(int, input().split())

result = solve(a_triplet, b_triplet)
alice, bob = map(str, result)

print("Alice: {0}, Bob: {1}".format(alice, bob))
```

```python
def solve(a_triplet, b_triplet):
    alice_points = 0
    bob_points = 0
    return [alice_points += 1 if a_val < b_val elif a_val > b_val bob_points += 1 for a_val, b_val in zip(a_triplet, b_triplet)]

a_triplet = map(int, input().split())
b_triplet = map(int, input().split())

result = solve(a_triplet, b_triplet)
alice, bob = map(str, result)

print("Alice: {0}, Bob: {1}".format(alice, bob))
```
