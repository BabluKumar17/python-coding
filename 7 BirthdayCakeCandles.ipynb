```python
from collections import Counter

def birthdayCakeCandles(n, arr):
    if n == len(arr):
        return  max(Counter(arr).items(), key=lambda item: item[0])

    
    
    
# number of candles
n = int(input().strip())

# heights of candles
arr = list(map(int, input().strip().split(' ')))

result = birthdayCakeCandles(n, arr)
print(result)


####### Sanity check #########
print(birthdayCakeCandles(5, [1, 2, 2, 3, 3]))               # (3, 2)
print(birthdayCakeCandles(9, [1, 2, 2, 3, 3, 3, 2, 1, 4]))   # (4, 1)
```
